```python
import numpy as np
import glob
import os
```

```python
def read_header_string(filename):
    header = ''
    header_length = 0
    with open(filename) as PMT_file:
        line = ''
        while not ("End of Header" in line):
            line = next(PMT_file)
            header += line
            header_length += 1
    return header_length, header[:-1]
```

```python
data_folder = 'O:\\Nat_IFA-fullc\\experimental\\expdata\\Fluorescence_LUNA2\\PMTdata\\2022\\15-12-2022'
save_folder = 'O:\\Nat_IFA-fullc\\experimental\\expdata\\Fluorescence_LUNA2\\PMTdata_summed\\2022\\15-12-2022'
os.listdir(data_folder)
```

```python
for run in os.listdir(data_folder):
    data_files = glob.glob(data_folder+'\\'+run+'\\*.dat')
    print(run)
    for file in data_files:
        header_length, header = read_header_string(file)
        data = np.loadtxt(file, skiprows=header_length)
        summed_data = np.sum(data, axis=1)
        sum_filename = save_folder + '\\'+run+'\\'+file.rsplit('\\', 1)[1][:-4]+'_sum.dat'
        print(sum_filename)
        np.savetxt(sum_filename, summed_data, header=header, comments='')
        # break
```
